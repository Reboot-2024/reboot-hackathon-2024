```python
import pandas as pd
import math
import re
import numpy as np
```

```python
df_manual = pd.read_csv('Manual Labelling - Sheet1.csv')
```

```python
def format_phone(stng):
    if isinstance(stng, str):
        flag = False
        if '+' in stng:
            flag = True
        numbers = re.findall(r'\d+', stng)
        numbers = ''.join(numbers)
        if flag:
            numbers = "+"+numbers
        return str(numbers)
    return np.nan
```

```python
import re
import numpy as np
def get_numbers(text):
    text = "Hello 123 world 456"
    numbers = re.findall(r'\d+', text)
    numbers = ''.join(numbers)
    return numbers
```

```python
def extract_experience(text):
  match = re.search(r'\d+(\.\d+)?', str(text))
  if match:
    return (float(match.group()))
  else:
    return np.nan
```

```python
def format_value(value):
  if isinstance(value, (int, float)) and value.is_integer():
    return str(int(value))
  else:
    return str(value)
df_manual['Total Experience2'] = df_manual['Total Experience'].apply(format_value)
df_manual['Phone Number2'] = df_manual['Phone Number'].apply(format_phone)
# df_manual['Estimated Matching Percentage2'] = df_manual["Estimated Matching Percentage"].apply(format_value).apply(extract_experience)
df_manual
```

```python
def percentage_str(text):
  match = re.search(r'\d+(\.\d+)?', str(text))
  if match:
    return (int(math.ceil(float(match.group()))))
  else:
    return np.nan
```

```python
def accuracy_str(ID, manual, test):
    stng = {}
    score = 0
    for key,x,y in zip(ID, manual, test):
        if isinstance(x, str) and isinstance(y, str):
            x = ' '.join(x.split())
            y = ' '.join(y.split())
            if x.lower()==y.lower():
                stng[key] = 1
                score +=1
            else:
                print(x,' mismatches ',y)
                stng[key] = 0
        elif isinstance(x, float):
            if math.isnan(x):
                stng[key] = 1
                score +=1
        elif isinstance(y, float):
            if math.isnan(y):
                stng[key] = 0
                print(x, ' mismatches ', y)
    return (score/len(stng)) *100
```

```python
def accuracy_phone(ID, manual, test):
    number = {}
    score = 0
    for key,x,y in zip(ID, manual, test):
        if isinstance(x, str) and isinstance(y, str):
            x = ' '.join(x.split())
            y = ' '.join(y.split())
            if str(x).lstrip("+")==str(y).lstrip("+"):
                number[key] = 1
                score +=1
            else:
                print(x,' mismatches ',y)
                number[key] = 0
        elif isinstance(x, float):
            if math.isnan(x):
                number[key] = 1
                score +=1
        elif isinstance(y, float):
            if math.isnan(y):
                number[key] = 0
                print(x, ' mismatches ', y)
    return (score/len(number)) *100
```

## Percentage Matching

```python
def accuracy_percentage(ID, manual, test):
    matching_percentage = {}
    score = 0

    for key,x,y in zip(ID, manual, test):
        difference = abs(x - y)
        
        # Introduce a 5% tolerance
        tolerance = 0.25 * max(x, y)
        
        if difference <= tolerance:
            matching_percentage[key] = 1
            score += 1
            # print(x, ' matches ', y)
        else:
            print(x, ' mismatches ', y)
            matching_percentage[key] = 0

    accuracy_percentage = (score/len(matching_percentage)) *100
    return (accuracy_percentage) 
```

```python
acc_percentage = accuracy_percentage(df_manual['Resume Name'],df_manual['Estimated Matching Percentage'],df_streamlit['Matching Percentage2'])
print(acc_percentage)
```
